# 2. Data Wrangling
### 데이터를 처음 format에서 분석하기 좋은 형태로 바꾸는 것

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
# 데이터 불러오기
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders","engine-size", "fuel-system","bore", "stroke", "impression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
car = pd.read_csv(url, names=headers)
car
```

### Convert ? to NaN
? : Python's default missing value marker, 속도와 편의성 때문

```python
# ?를 NaN으로 바꾸기 - replace
car.replace('?', np.nan, inplace = True )
car.head(3)
```

```python
# missing data 찾기 1
missing_data = car.isnull()
missing_data
# 2
missing_data = car.notnull()
missing_data
```

```python
missing_data.columns                     # index로 불러옴
missing_data.columns.values              # array로 불러옴
missing_data.columns.values.tolist()     # list로 불러옴
```

```python
# 컬럼별 missing value의 개수 세기
for column in missing_data.columns.values.tolist() :
    print(column)       # pandas의 dataframe아니고는 print 써야함
    print(missing_data[column].value_counts())     # 값을 보여줌
    print('')           # 한줄 띄우기
```

```python
# Normalized-loss열 NaN값을 mean으로 바꾸기
# dtype이 int64(평균 못 구함)라서 float으로 바꿔야 mean 계산 가능
# 컬럼만 부르면 series이기 때문에 1행밖에 없음 - axis 지정 안해도 됨
avg_norm_loss = car['normalized-losses'].astype('float').mean(axis=0)
avg_norm_loss
```

```python
# NaN을 평균값으로 바꾸기
car['normalized-losses'].replace(np.nan, avg_norm_loss, inplace=True)
car['normalized-losses'].values
```

```python
# Bore열 NaN값 mean으로 바꾸기
avg_bore = car['bore'].astype('float').mean()
avg_bore
# NaN값 바꾸기
car['bore'].replace(np.nan, avg_bore, inplace=True)
car['bore'].values
```

```python
# Stroke열 NaN값 평균으로 바꾸기
avg_stroke = car['stroke'].astype('float').mean()
avg_stroke
# NaN값 바꾸기
car['stroke'].replace(np.nan, avg_stroke, inplace=True)
car['stroke'].values
```

```python
# Horespower의 NaN값 바꾸기
avg_hpw = car['horsepower'].astype('float').mean()
avg_hpw
car['horsepower'].replace(np.nan, avg_hpw, inplace=True)
```

```python
# Peak-rpm의 NaN값 바꾸기
avg_rpm = car['peak-rpm'].astype('float').mean()
avg_rpm
car['peak-rpm'].replace(np.nan, avg_rpm, inplace=True)
```

```python
# Num-of-doors의 NaN값 바꾸기 - frequency(mode)로 바꾸기
car['num-of-doors'].value_counts()            # 특정 컬럼의 값 보기
car['num-of-doors'].value_counts().idxmax()   # 그 중 최빈도 구하기
car['num-of-doors'].replace(np.nan, car['num-of-doors'].value_counts().idxmax(), inplace=True)
# 또는
car['num-of-doors'].replace(np.nan, "four", inplace=True)
car['num-of-doors'].value_counts()
```

```python
# price(구하려는 값)의 NaN값이 있는 행은 drop하기
car.dropna(subset=['price'], axis=0, inplace=True)
car['price'].isnull().value_counts()
```

# Data Normalization

자료 분석할 때(such as linear regression), 자료의 단위가 큰 쪽이 단위가 작은쪽보다 impact가 크다.

이것을 피하기 위해 모두 eg) 0~1 사이의 값으로 변환하는 것 : normalization

**Normalizaion 방법**
* Simple feature scaling(0~1) : 그 열의 최대값으로 나눠주는 것
* Min-Max(0~1) : 그 열의 최소값과의 차이를 범위차이로 나눠주는 것, x_new = (x_old -x_min)/(x_max - x_min)
* Z-score(보통 -3~3) : 표준정규분포로 전환

```python
# Simple feature scaling으로 normalization
car['length'] = car['length']/car['length'].max()
car['width'] = car['width']/car['width'].max()
car['height'] = car['height']/car['height'].max()
car[['length', 'width', 'height']]
```

```python
# Min-Max로 normalization
(car['highway-mpg']-car['highway-mpg'].min()) / (car['highway-mpg'].max()-car['highway-mpg'].min())
```

```python
# Z-score normalization
(car['highway-mpg']-car['highway-mpg'].mean()) / car['highway-mpg'].std()
```

# Binning in Python
numerical을 categorical 'bin'으로 바꾸어서 그룹분석을 위한 것

-->  예측모델의 정확도를 높여줄 수 있음

```python
# highway-mpg를 3구간(low, medium, high)으로 나누기 --> 4개의 divider가 필요함
bins = np.linspace(min(car['highway-mpg']), max(car['highway-mpg']), 4)      # 구간자르기
group_names = ['low', 'medium', 'high']                                      # 구간 이름
car['highway-mpg_binned'] = pd.cut(car['highway-mpg'], bins, labels=group_names)
```

```python
# Horsepower의 datatype 변경
car['horsepower'] = car['horsepower'].astype('int64', copy=True)

# Histogram으로 distribution 확인하기
import matplotlib as plt
from matplotlib import pyplot
%matplotlib inline
plt.pyplot.hist(car['horsepower'])    # histogram 만들 대상

# plot label 붙이기
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
# Bin 만들기
bins = np.linspace(car['horsepower'].min(), car['horsepower'].max(), 4)
bins
```

```python
# Bin names
bin_names = ['Low', 'Medium', 'High']

# horsepower를 binning하기
car['horsepower-binned'] = pd.cut(car['horsepower'], bins, labels=bin_names, include_lowest=True)
car[['horsepower', 'horsepower-binned']].head(30)

# 각 구간의 개수 보기
car['horsepower-binned'].value_counts()
```

```python
# Bar graph로 Plotting
pyplot.bar(group_names, car['horsepower-binned'].value_counts())
plt.pyplot.xlabel('horsepower-group')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
# Histogram으로 plotting
plt.pyplot.hist(car['horsepower'], bins=3)
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('hosepower bins')
```

# Turning categorical variables into quantitative variables in Python
통계모델에 넣기 위해서는 문자이면 안되는 경우가 대부분

Categorical --> Numeric (indicator/dummy variable) : Onehot encoding --> Regression analysis 위해서 
* <b>Indicater/Dummy variable</b> : category를 numeric variable로 바꿔주기 위한 것, 숫자가 의미하는 바가 없기 때문에 "dummies"

```python
# fuel-type 행만 인코딩됨
encoded = pd.get_dummies(car['fuel-type'])
encoded.head()
```

```python
# 정확하게 하기 위해 컬럼 이름 바꾸기
encoded.rename(columns={'diesel':'fuel-type-diesel', 'gas':'fuel-type-gas'}, inplace=True)
encoded
car
```

```python
# 먼저 원래 dataframe에 위에 만들어진 인코딩 dataframe을 합치기 - concatenate
df = pd.concat([car, encoded], axis=1)
# fuep-type 컬럼 drop
df.drop(columns=['fuel-type'], inplace=True)
# 또는
df.drop('fuel-type', axis=1, inplace=True)
```

```python
# Aspiration 컬럼에도 해보기
df['aspiration'].value_counts()     # std, turbo로 구성

# onehot-encoding : dummy 만들기
encoded = pd.get_dummies(car['aspiration'])

# 기존 테이블에 합치기
df1 = pd.concat([df, encoded], axis=1)

# 이름 바꿔주기
df1.rename(columns={"std":"aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)

# 기존 aspriation 컬럼 drop
df1.drop('aspiration', axis=1, inplace=True)
df1
```
